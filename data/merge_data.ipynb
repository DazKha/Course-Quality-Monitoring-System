```python
import pandas as pd

data_dir = "/Users/minhkha/Desktop/DS317/course_quality_monitor/data"

# Chỉ lấy cột id và name để nhẹ bộ nhớ
course = pd.read_csv(
    f"{data_dir}/course.csv",
    usecols=["id", "name"]
)

train = pd.read_csv(f"{data_dir}/train_set.csv", low_memory=False)

merged = (
    train
    .merge(course, how="left", left_on="course_id", right_on="id")
    .rename(columns={"name": "course_name"})
    .drop(columns=["id"])
)

```

```python
import pandas as pd

# Đọc course_score chỉ lấy cột cần thiết
course_score = pd.read_csv(
    f"{data_dir}/course_score.csv",
    usecols=["course_id", "CQV", 'CQS'],
    low_memory=False,
)

merged = (
    train
    .merge(course, how="left", left_on="course_id", right_on="id")
    .merge(course_score, how="left", on="course_id")
    .rename(columns={"name": "course_name"})
    .drop(columns=["id"])
)

merged.to_csv(f"{data_dir}/train_set_with_name_score.csv", index=False)

# cols_for_avg = [
#     "assignment_coverage",
#     "video_coverage",
#     "discussion_coverage",
#     "n_users_content_interaction",
#     "correct_rate_course",
#     "progress_ratio",
# ]

# # Tính trung bình theo hàng, bỏ qua NaN
# merged["learning_interaction_score"] = merged[cols_for_avg].mean(axis=1, skipna=True)
```

```python
import pandas as pd
import glob

data_dir = "/Users/minhkha/Desktop/DS317/course_quality_monitor/data"

# Đọc course để lấy name
course = pd.read_csv(
    f"{data_dir}/course.csv",
    usecols=["id", "name"]
)

# Lấy tất cả các file CSV trong thư mục predicted
predicted_files = glob.glob(f"{data_dir}/predicted/*.csv")

for file_path in predicted_files:
    # Đọc file predicted
    df = pd.read_csv(file_path, low_memory=False)
    
    # Merge với course để thêm course_name
    df_with_name = (
        df
        .merge(course, how="left", left_on="course_id", right_on="id")
        .rename(columns={"name": "course_name"})
        .drop(columns=["id"])
    )
    
    # Lưu lại file (ghi đè hoặc tạo file mới)
    df_with_name.to_csv(file_path, index=False)
    print(f"Đã thêm course_name vào: {file_path}")
```
